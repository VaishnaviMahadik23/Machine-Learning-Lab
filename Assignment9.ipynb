```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
n_samples = 300
n_features = 2  # 2 dimensions for easy visualization
n_clusters = 4  # We'll try to find 4 clusters
```

```python
X, y_true = make_blobs(n_samples=n_samples, centers=n_clusters,
                       cluster_std=0.60, random_state=42)
```

```python
print(f"Shape of the generated data (X): {X.shape}")
print("-" * 30)
```

```python
# --- 2. Initialize and Train the K-Means Model ---

# Define the number of clusters (K)
k = n_clusters
```

```python
model = KMeans(n_clusters=k, random_state=42, n_init='auto')
```

```python
model.fit(X)
```

```python
y_predict = model.predict(X)
```

```python

centers = n_clusters

print(f"Cluster Centers (Coordinates):\n{centers}")
print("-" * 30)
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(X[:, 0], X[:, 1], c=y_predict, s=50, cmap='viridis', label='Data Points')

centers = model.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.9, marker='X',
            label='Cluster Centers')

plt.title(f"K-Means Clustering with K={k}")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid(True)
plt.show()

print(f"First 10 assigned cluster labels: {y_predict[:10]}")
```
