```python
upto = int(input("Find sum of prime numbers upto: "))

sum_primes = 0

for num in range(2, upto + 1):  
    is_prime = True
    for i in range(2, int(num**0.5) + 1):  
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        sum_primes += num

print("\nSum of all prime numbers upto", upto, ":", sum_primes)
```
